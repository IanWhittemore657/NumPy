## Indexing a 2D array (matrices)

```python
arr_2d = np.array(([5,10,15],[20,25,30],[35,40,45]))

#Show
arr_2d
```

```python
#Indexing row
arr_2d[1]
```

```python

# Getting individual element value
arr_2d[1][0]
```

```python
#does the same thing as above just a different format
arr_2d[1,0]
```

```python
# 2D array slicing
arr_2d[:2,1:]
```

```python
#Shape bottom row
arr_2d[2]
```

```python
#Shape bottom row
arr_2d[2,:]
```
